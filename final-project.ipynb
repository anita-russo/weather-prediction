# USING BOKEH LIBRARY
NOTE: install bokeh first using 'pip install bokeh' on cmd before running

```python
df = pd.read_csv('Toronto_temp.csv')
```

```python
from bokeh.plotting import figure
from bokeh.io import output_file, show
from bokeh.models import SingleIntervalTicker, LinearAxis

output_file('average-temperatures.html')
f = figure(x_axis_type=None)

monthly_mean = df.groupby('Month').mean()
months = monthly_mean.index
averages = monthly_mean['Mean Temp (C)']

#get all temperatures
temps = []
for i in range(1,13):
    temps.append(df[df['Month'] == i]['Mean Temp (C)'])

#add temperatures to graph
c = 0
for i in temps:
    f.circle(c+1, temps[c],size=1,legend='Temperatures')
    c+=1
    
f.yaxis.axis_label = 'Temperature (C)'
f.add_layout(LinearAxis(ticker=SingleIntervalTicker(interval=1)), 'below')
f.xaxis.major_label_overrides = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun',
                                 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}

f.line(months, averages, color="red", line_dash=[4, 10],legend="Average")

show(f)
```

# USING SCIKITLEARN

```python
import sklearn
from sklearn import tree
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
#drop null values & split training and testing data
df.dropna(how='any',inplace=True)
train, test = train_test_split(df, test_size=0.2)
```

```python
#get training data
feature_cols = ['Month', 'Day']
X = train.loc[:,feature_cols]
Y = train['Mean Temp (C)'].astype(int)
```

```python
#create classification model
logreg = LogisticRegression(solver='lbfgs')
logreg.fit(X,Y)
```

```python
#use testing data
X_new = test.loc[:,feature_cols]
new_pr = logreg.predict(X_new)
```

```python
#generate dataframe of prediction & actual values
pred = pd.DataFrame(X_new)
pred['Prediction'] = new_pr
pred['Actual'] = test.loc[:,'Mean Temp (C)']
```
